```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
K = tf.keras.backend
```

```python
data=pd.read_csv('../data/data_bert.csv',encoding='utf-8')
```

```python
train_data = data.sample(frac=1)
train_data.head()
```

```python
data.head()
```

```python
X_data=data.drop(['lable','Unnamed: 0'],axis=1)
```

```python
X_data_array=X_data.values
```

```python
len(X_data_array[0])
```

```python
y=np.array(data["lable"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_array, y, test_size=0.2,
                                                        random_state=2022)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=64,
                                input_dim=1542,
                                use_bias=True,
                                kernel_initializer='uniform',
                                bias_initializer='zeros',
                                activation='relu'))
model.add(tf.keras.layers.Dense(units=32,
                                activation='relu'))
model.add(tf.keras.layers.Dense(units=1,
                                activation='sigmoid'))
```

```python
model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(0.003),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, epochs=100, batch_size=512, validation_data=(X_test, y_test))
    
```

```python
y_p=model.predict(X_test)
```

```python
model.save('bert-DNN.h5')
```
