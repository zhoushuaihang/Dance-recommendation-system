```python
import pandas as pd
```

```python
dance=pd.read_csv('data/dance_bert_em.csv',encoding='utf-8')
```

```python
dance=dance.drop(['Unnamed: 0'],axis=1)
dance.head()
```

```python
action=pd.read_csv('data/action.csv')
```

```python
len(action)
```

```python
user=pd.read_csv('data/user_em.csv',encoding='utf-8')
```

```python
user=user.drop(['Unnamed: 0','birthday'],axis=1)
user
```

```python
data_t=pd.merge(action,user,on=['user_id'])
```

```python
data_t=data_t.drop(['Unnamed: 0'],axis=1)
```

```python
data_t
```

```python
data=pd.merge(data_t,dance,on=['dance_id'])
```

```python
data.head()
```

```python
data.to_csv("data/data_bert.csv",encoding="utf-8")
```
