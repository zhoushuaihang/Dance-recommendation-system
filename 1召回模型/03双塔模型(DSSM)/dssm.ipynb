```python
num_users=data["user_nm"].max()+1
num_dances=data["dance_nm"].max()+1
num_level=data["level"].max()+1
num_sex_em=data["sex_em"].max()+1
num_favorites=data["favorites"].max()+1
num_play=data["play"].max()+1
```

```python
#构建训练数据集
# df_sample=data.sample(frac=0.1)
X=data[["user_nm","level","sex_em","dance_nm","favorites","play"]]

y=data.pop("lable")
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                        random_state=2022)
```

```python
X_train
```

```python
#搭建双塔模型并训练
def get_model():
    """函数式API搭建双塔DNN模型"""
    #输入
    user_nm=keras.layers.Input(shape=(1,),name="user_nm")
    level=keras.layers.Input(shape=(1,),name="level")
    sex_em=keras.layers.Input(shape=(1,),name="sex_em")
    favorites=keras.layers.Input(shape=(1,),name="favorites")
    play=keras.layers.Input(shape=(1,),name="play")
    dance_nm=keras.layers.Input(shape=(1,),name="dance_nm")
    
    #user塔
    user_vector=tf.keras.layers.concatenate([
        layers.Embedding(int(num_users),100)(user_nm),
        layers.Embedding(num_level,2)(level),
        layers.Embedding(num_sex_em,2)(sex_em) 
        
    ])
#     user_vector=layers.Dense(512,activation='relu')(user_vector)
#     user_vector=layers.Dropout(0.5)(user_vector)
    user_vector=layers.Dense(256,activation='relu')(user_vector)
#     user_vector=layers.Dropout(0.5)(user_vector)
    user_vector=layers.Dense(128,activation='relu')(user_vector)
    user_vector=layers.Dropout(0.2)(user_vector)
    user_vector=layers.Dense(32,activation='relu',name="user_embedding",kernel_regularizer="l2")(user_vector)
    
    #dance塔
    dance_vector=tf.keras.layers.concatenate([
        layers.Embedding(int(num_dances),100)(dance_nm),
        layers.Embedding(num_favorites,2)(favorites),
        layers.Embedding(num_play,2)(play)
    
        
    ])
#     dance_vector=layers.Dense(512,activation='relu')(dance_vector)
#     dance_vector=layers.Dropout(0.5)(dance_vector)
    dance_vector=layers.Dense(256,activation='relu')(dance_vector)
#     dance_vector=layers.Dropout(0.5)(dance_vector)
    dance_vector=layers.Dense(128,activation='relu')(dance_vector)
    dance_vector=layers.Dropout(0.2)(dance_vector)
    dance_vector=layers.Dense(32,activation='relu',name="dance_embedding",kernel_regularizer="l2")(dance_vector)
    #合并
    user_dance=tf.reduce_sum(user_vector*dance_vector,axis=1)
    user_dance=tf.expand_dims(user_dance,1)
    #模型输出
    output_model=layers.Dense(1,activation="sigmoid")(user_dance)
    
    return keras.models.Model(inputs=[user_nm,level,sex_em,dance_nm,favorites,play],outputs=(output_model))
    
```

```python
DSSM=get_model()
```

```python
# DSSM.compile(loss=tf.keras.losses.MeanSquaredError(),optimizer=keras.optimizers.RMSprop())
```

```python
DSSM.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
fix_x_train=[
    X_train['user_nm'],
    X_train['level'],
    X_train["sex_em"],
    X_train["dance_nm"],
    X_train["favorites"],
    X_train["play"]  
]
from datetime import datetime
TIME="{0:%Y-%m-%dT%H-%M-%S/}".format(datetime.now())
tensorboard_callback=tf.keras.callbacks.TensorBoard(log_dir='./logs/logs_'+TIME)
```
