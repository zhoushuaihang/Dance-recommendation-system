```python
data=pd.read_csv('data/data_em.csv')
```

```python
len(data)
```

```python
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
X_data=data.drop(['lable'],axis=1)
```

```python
X_data_array=X_data.values
```

```python
y=np.array(data["lable"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_array, y, test_size=0.2,
                                                        random_state=2022)
```

```python
YoutobeDNN = tf.keras.models.Sequential()
YoutobeDNN.add(tf.keras.layers.Dense(units=512,
                                input_dim=606,
                                use_bias=True,
                                kernel_initializer='uniform',
                                bias_initializer='zeros',
                                activation='relu'))
YoutobeDNN.add(tf.keras.layers.Dense(units=256,
                                activation='relu'))
YoutobeDNN.add(tf.keras.layers.Dense(units=128,
                                activation='relu'))
YoutobeDNN.add(tf.keras.layers.Dense(units=32,
                                activation='relu'))
YoutobeDNN.add(tf.keras.layers.Dropout(rate=0.5))
YoutobeDNN.add(tf.keras.layers.Dense(units=1,
                                activation='sigmoid'))
```

```python
YoutobeDNN.summary()
YoutobeDNN.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history=YoutobeDNN.fit(X_train, y_train, epochs=100, batch_size=256, validation_data=(X_test, y_test))
```

```python
YoutobeDNN.save("myYoutobeDNN.h5")
```
