```python
import jieba
import re
```

```python
wlist=[]
for word in wordlist:
    word=str(word)
    r1 = "[a-zA-Z0-9\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、：；;《》“”~@#￥%……&*（）]+"
#     word=re.sub(r1,'',word)
    word = re.sub(r1, '',word)
    wlist.append(word)
```

```python
def cut_stop(s):
    stop_list =['「','」','‖','【','】','🎵','🤩','～','°͜ʖ͡°','✧',')',
                '-','|','♥',' ','｜','•','❀','☽','〕','〔','❤️','♪','♬',
                '˘','︶','♡',':','『','☞','✌️','🌿','😂','😎','』','］',
                '［',']','[','°','✾','✿','〖','〗','☺️','?','㊗️',"ʖ͡'",
                '❤️','✨','🦋','·','×','⑦','>',"｀･∧･´",'з','∠',
                '丨',"≧﹏≦",'×','▷','▶','◀','◁','❤','🌕','·','❤️','㊗️','✌️','☺️','💫']
    words =''
    for word in s:
        if word not in stop_list:
            words += word
        else:
            words += ""
    return words
```

```python
word_list=[]
for w in wlist:
    word_list.append(cut_stop(w))
# print(word_list)  
```

```python
import pickle
with open('word.pickle','wb') as f:
    pickle.dump(word_list, f)
```

```python
lines = []
for line in word_list: #分别对每段分词
    temp = jieba.lcut(line)  #结巴分词 精确模式
    words = []
    for i in temp:
        #过滤掉所有的标点符号
        i = re.sub("[\s+\.\!\/_,$%^*(+\"\'””《》]+|[+——！，。？、~@#￥%……&*（）：；‘]+", "", i)
        if len(i) > 0:
            words.append(i)
    if len(words) > 0:
        lines.append(words)
print(lines[0:5])#预览前5行分词结果
```
