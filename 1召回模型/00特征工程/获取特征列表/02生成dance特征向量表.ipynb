```python
def extract_word(wordlist):
    wlist=[]
    word=str(wordlist)
    r1 = "[a-zA-Z0-9\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、：；;《》“”~@#￥%……&*（）]+"
    #     word=re.sub(r1,'',word)
    word = re.sub(r1, '',word)
    wlist.append(word)
    return wlist
```

```python
print(extract_word(title[1]))
```

```python
def cut_stop(s):
    stop_list =['「','」','‖','【','】','🎵','🤩','～','°͜ʖ͡°','✧',')',
                '-','|','♥',' ','｜','•','❀','☽','〕','〔','❤️','♪','♬',
                '˘','︶','♡',':','『','☞','✌️','🌿','😂','😎','』','］',
                '［',']','[','°','✾','✿','〖','〗','☺️','?','㊗️',"ʖ͡'",
                '❤️','✨','🦋','·','×','⑦','>',"｀･∧･´",'з','∠',
                '丨',"≧﹏≦",'×','▷','▶','◀','◁','❤','🌕','·','❤️','㊗️','✌️','☺️','💫']
    words =''
    for word in s:
        if word not in stop_list:
            words += word
        else:
            words += ""
    return words
```

```python
print(cut_stop(title[1]))
```

```python
def cutword(word_list): 
    temp = jieba.lcut(word_list)  #结巴分词 精确模式
    words = []
    for i in temp:
        #过滤掉所有的标点符号
        i = re.sub("「」‖[\s+\.\!\/_,$%^*(+\"\'””《》]+|[+——！，。？、~@#￥%……&*（）：；‘]+「", "", i)
        if len(i) > 0:
            words.append(i)
#     if len(words) > 0:
#         lines.append(words)
#     print(lines[0:5])#预览前5行分词结果
    return words
```

```python
def finalcut(word_list):
    word=extract_word(word_list)
    wlist=cut_stop(word)
    wordlist=cutword(wlist)
    return wordlist
```

```python
print(finalcut(title[1]))
```

```python
from gensim.models import KeyedVectors
def load_word2vec_model(w2v_path):
    # load word2vec
    model =KeyedVectors.load_word2vec_format(w2v_path, binary=True)
    return model
model = load_word2vec_model('Word2Vec60.bin')
```

```python
def getwordvec(word_list):
    list_title=[]
    for i in finalcut(word_list):
        try:
            list_title.extend(model.get_vector(i))
        except Exception as E:
            continue
    return list_title
```
